# Web scraping SFC enforcement News

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import numpy as np
import time

URL = 'https://apps.sfc.hk/edistributionWeb/gateway/EN/news-and-announcements/news/enforcement-news/'

page = requests.get(URL)

soup = BeautifulSoup(page.content,'html.parser')

results = soup.find('div', class_= lambda text:'enforcement-news-table'.lower())

path = r'F:\\AI_ML\\Software\\chromedriver_win32\\Chromedriver'

driver = webdriver.Chrome(executable_path = path)

driver.get('https://apps.sfc.hk/edistributionWeb/gateway/EN/news-and-announcements/news/enforcement-news/')
```

```python
rootelement = driver.find_element_by_id('root')
#print(rootelement)
datatable = []
theads = []
time.sleep(3)
roottable = rootelement.find_element_by_tag_name("table")
print(roottable)
#returns thead list
thlist = roottable.find_elements_by_tag_name("th")
for th in thlist:
    #returns sub element list anchor list
    aas = th.find_elements_by_tag_name("a")
    for anc in aas:
    #anchor text property extracted
        thead = anc.get_property("text")
    #th row data
        theads.append(thead)

theads.append('URL')
# theads.append('heading')
# theads.append('content')
```

```python
#adding URL Column to thead

#print('------data----')
#getting tbody element not a list element method is changed
tbodys = roottable.find_elements_by_tag_name("tbody")
#second tbody is holding data trs
trs = tbodys[1].find_elements_by_tag_name("tr")
for tr in trs:
    data=[]
    tds = tr.find_elements_by_tag_name("td")
    index=0
    for td in tds:
        value = td.get_property("innerText")
        data.append(value)
        if(index==2):
            anchor=td.find_element_by_tag_name("a")
            url = anchor.get_property("href")
            data.append(url)
        index=index+1
    
    #adding data to the data table
    datatable.append(data)
        
#to dataframe
df = pd.DataFrame(datatable)
#setting columns
df.columns = theads

df.head()
```

```python
articles = []
# article.append('heading')
# article.append('content')
def my_function(url):
    print(url)
    driver.get(url)
    time.sleep(5)
    divcontent = driver.find_element_by_id('content')

    headingElement = divcontent.find_element_by_tag_name('h1')
    heading = headingElement.get_property('innerText')

    newsdiv = divcontent.find_element_by_class_name('newsc')
    paras = newsdiv.find_elements_by_tag_name('p')
    content = ''
    for para in paras:
        #print('\n'*2, para.get_property('innerText'))
        content = content + para.get_property('innerText') + '\n'*2
    
    return heading, content
# index = 0
# for index, url in df["URL":]:
#     print(url)
```

```python
headinglist = []
contentlist = []

for url in df["URL"]:
    heading, content =  my_function(url)
    print('heading: ',heading)
    headinglist.append(heading)
    #print('\ncontent:\n',content)
    contentlist.append(content)
    time.sleep(3)
df1 = df.copy()    
df1['heading'] = pd.Series(headinglist, index = df1.index)
df1['content'] = pd.Series(contentlist, index = df1.index)
df1.head()
     
```
